```python
dd_deptos = pd.read_csv('data/dd_deptos.csv', sep=';', decimal=',')
print(dd_deptos.shape)
dd_deptos.head(2)
```

```python
def aumentar_edad_promedio(n):
    "Recalcula edad promedio"
    edad_prom_des_new = dd_deptos.edad_prom_des * n
    r = np.log(edad_prom_des_new)

    return r
```
